#Importing datasets from github

To import any dataset from github do the following steps - 
1. Search for the dataset you want on github.
2. Once you are on the page of your desired datset click on the RAW tab on the screen. as shown below - 

```
from IPython.display import Image
Image("c1.PNG")
```

3. Copy the URL and store it in variable named URL (demonstrated in this notebook).
4. Now use this URL in your notebook directly. as shown below - 

## Dataset 1: Bene_insurance.csv 

```
import pandas as pd
URL = "https://raw.githubusercontent.com/DSpython/importing_datasets/master/Datasets/Bene_insurance.csv"
df = pd.read_csv(URL)
df.head(2)
```
